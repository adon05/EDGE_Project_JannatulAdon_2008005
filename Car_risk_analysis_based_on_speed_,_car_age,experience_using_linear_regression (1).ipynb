```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('car risk.csv')
```

```python
df
```

```python
df.experience
```

```python
exp_fit = df.experience.median()
exp_fit
```

```python
df.experience = df.experience.fillna(exp_fit)
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['speed','car_age','experience']],df['risk'])
```

```python
reg.predict([[160,10,5]])
```

```python
df['pridicted risk']= reg .predict(df[['speed','car_age','experience']])
df
```

```python
plt.plot(df['risk'])
plt.plot(df['pridicted risk'], color = 'red')
```

```python
r_squared = reg.score(df[['speed', 'car_age', 'experience']], df['risk'])

print("R-squared:", r_squared)
```
